## What makes a good wine?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
wine = pd.read_csv('winequality-red.csv')
wine.head()
```

```python
sns.heatmap(wine.corr(), linewidths=0.5, square= True)
plt.show()
```

```python
wine.corr()['quality'].abs().sort_values(ascending=False)
```

```python
top_three_corr = wine.corr()['quality'].abs().sort_values(ascending=False).index[1:4]
```

```python
wine['quality'].value_counts().plot.bar()
plt.show()
```

```python
for i in wine.columns:
    sns.countplot(wine[i])
    plt.show()
```

```python
for i in top_three_corr:
    sns.boxplot(x='quality', y=i, data=wine)
    plt.show()
```

```python
sns.jointplot(x='chlorides', y='quality', data=wine,
             kind='reg')
```

```python
sns.jointplot(x='alcohol', y='quality', data=wine,
             kind='reg')
```

```python
wine['quality'].astype('category')
```

```python
sns.jointplot(x='volatile acidity', y='quality', data=wine,
             kind='reg')
```
